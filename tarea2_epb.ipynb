EJERICIO 1: GENERAR FUNCIÓN QUE DIGA HEXAGONO

```python
#EJERCICIO 1 DEVOLVER UN HEXAGONO: 3 INTENTOS.
#Primero genero un Data Frame con numeros aleatorios. Los valores de la tabla van del 1 al 5 y las filas son 10 con 3 columnas. 
np.random.seed(1234)
df1=pd.DataFrame(
    data=np.random.randint(1,5,(10,3)),
     columns=["A","B","C"]
     )
df1
```

```python
#Hacer un hexagono: ENTENDÍ ERRONEAMENTE QUE SERÍA POTENCIA. En este caso elegí la columna o variable "C" para convertirla en hexagono.
df1["hexa"]=df1['C']**6
print(df1)
```

```python
#HEXAGONO: COLUMNA DE SEIS LADOS. CREAR DIVERSAS COLUMNAS CON DISTINTAS ITERACIONES: Aca devuelvo una función en una nueva columna "D" que me indique si la suma de las columnas B+C=6, entonces hace referencia a un área de 6 lados o hexágono
#Paso 1, copio en df1 y creo un df2. 
df2=df1.copy()
df2["D"]=df1["B"] + df1["C"]
df2
```

```python
#Posteriormente se anexa la nueva columna "E", donde 6 hace referencia a un objeto con 6 lados u hexagono. "Lambda es para ejecutar una función anónima"
df2["E"]=df2.apply(lambda x: "hexagono" if x.D == 6 else "No es hexagono", axis=1)
df2
                   
```

```python
#Intento 3: FIGURA . Hacer una figura con ciclos. Es una manera artesanal para hacer ciclo. "Rango" lo que hace es devolver una serie de valores (el ciClo se va a ejecutar n veces que tenga el rango)
#  *** DOS ESPACIOS Y TRES ASTERISCOS
# ***** UN ESPACIO Y CINCO ASTERISCOS 
#******* SIN ESPACIO Y 7 ASTERISCOS 
# ***** UN ESPACIO Y CINCO ASTERISCOS
#  *** DOS ESPACIOS Y TRES ASTERISCOS

#Esto se puede realizar de dos maneras: MODO 1: crear un ciclo "for" con condicionales, pero antes que tiene que declarar a las lineas con los espacios que conforman el hexagono 

linea1='  '  #Son 2 espacios
linea2=' '   #Un espacio
linea3=''    #Sin espacio

for i in range(7): #Aca se va a crear un ciclo con un máximo de 7 asteriscos (rango), donde se crear tres diferentes lineas con distintas condicionales:
  if i <= 2:        #Aca estoy haciendo una condicional pidiendo una linea con tres asteriscos (3-1=2) . Porque i empieza en CERO Y TERMINA EN 6
    linea1=linea1+ '*'#Aca mando a llamar a la linea 1 que son "dos espacios o strings" y lo concateno con la cantidad de ***
  if i <= 4:          # Aca estoy haciendo una condicional pidiendo una linea con cinco asteriscos (5-1=4) 
    linea2=linea2+ '*' # Aca mando a llamar a la linea 2 que es "un string" y lo concateno con la cantidad de *****
  if i <= 6:          # Aca estoy haciendo una condicional pidiendo una linea con 7 asteriscos (7-1=6) 
    linea3=linea3+ '*' # Aca mando a llamar a la linea 3 que es "un string" y lo concateno con la cantidad de asteriscos condicionados *******
# Ya no hay nececidad de generar als lineas porque la parte final del hexagono ya fue creada.
print(linea1)
print(linea2)
print(linea3)
print(linea2)
print(linea1)

# MODO 2 Se generan las variables con espacios, que posteriormente serán contatenadas con los '*'
lineb1='  ' #Son 2 espacios
lineb2=' ' #1 espacio
lineb3='' #sin espacio

#Aca no se usan condicionales, simplemente se trae la linea creada con los strings o espacios, concatenandola con el número exacto se asteriscos que se piden. 
for i in range(3): 
    lineb1=lineb1+ '*' #el + es para concatenar cadenas. En este caso concatenamos espacios con el asterisco.    
for i in range(5):
    lineb2=lineb2+ '*' #Continuando el ciclo, se trae a 
for i in range(7):
    lineb3=lineb3+ '*'
print("========================")
print(lineb1)
print(lineb2)
print(lineb3)
print(lineb2)
print(lineb1)



```

```python
#Una vez imputada, checamos que no existan datos NaN o vacios. Para posteriormente pegar las variables que son string a esta DF.
#Pero antes, se tendrá que convertir la tabla imputada a DF, ya que posterior a la transformación, ésta quedo en "array". 
autos3 = pd.DataFrame(  
    data=autos2,
    columns=['Millas_Por_Galon', 'Cilindros', 'Peso_En_Libras', 'Desplazamiento', 'Aceleracion', 'Caballo_De_Fuerza'])
print(autos3)
```

```python
#Comprobación de que no hay datos Nan 
autos3.isnull().sum()
```

```python
#Generar col((umna con la siguiente formula: Log10(va1/var2) =log10. En este caso Va1=Aceleración y Var2=Cilindros
autos3["log10"]=np.log10(autos3['Aceleracion']/ autos3['Cilindros'])
print(autos3)
```

```python
#Generar una columna que sea:  Sqrt(var1)exp(var2)/200. Se crea la variable "Sqrt" que es la raiz cuadrada de "Aceleración" y que multiplica a la exponencial de "Cilindros"/200 

autos3["Sqrt"]=np.sqrt((autos3['Aceleracion']))*np.exp(autos3['Cilindros'])/200
print(autos3)
```

```python
#Generar una columna que cumpla, la siguiente condicional: Si var 1 > var 2 → 5, caso contrario 3. A esta columna la voy a llamar "Auto1", donde var1= Aceleración y var2=Cilindros
autos3["Auto1"]=autos3.apply(lambda x:5 if x.Aceleracion > x.Cilindros else "3", axis=1)
print(autos3)
```

```python
#Se crea la función de cajas y bigotes
def cajas_bigotes (columnas, data, x):
  for i, colum in enumerate (columnas):
    fig, ax = plt.subplots(figsize=(6, 3.84))
    sns.boxplot(
        x=x,
        y=colum,
        data=data,
        )
    ax.set_title(colum, fontsize = 10, fontweight = "bold")
    ax.tick_params(labelsize = 8)
    ax.set_xlabel("")
```
